# Limpieza de la base de datos

## Importar librerias

```python
import numpy as np
import pandas as pd
```

## Ingenieria de datos

```python
df = pd.read_csv('../database/tetuan-city-power-consumption.csv')
df.head()
```

```python
df.tail()
```

```python
df.columns
```

Renombrando las columnas

```python
df.rename(columns={
    'DateTime' : 'datetime',
    'Temperature' : 'temp',
    'Humidity' : 'humi',
    'Wind Speed' : 'ws',
    'general diffuse flows' : 'gdf',
    'diffuse flows' : 'df',
    'Zone 1 Power Consumption' : 'z1-pc',
    'Zone 2  Power Consumption' : 'z2-pc',
    'Zone 3  Power Consumption' : 'z3-pc',
}, inplace=True)

df.head()
```

Cambiando el formato de la columna `datetime`

```python
df['datetime'] = pd.to_datetime(df['datetime'], format='%m/%d/%Y %H:%M')
df.head()
```

Consultando la informacion general del dataframe

```python
df.info()
```

Extrayendo los campos de `year`, `month`, `day`, `hour`, `minute` del campo `datetime`

```python
df['year'] = df['datetime'].dt.year
df['month'] = df['datetime'].dt.month
df['day'] = df['datetime'].dt.day
df['hour'] = df['datetime'].dt.hour
df['minute'] = df['datetime'].dt.minute
df.head()
```

Consultando la informacion general del dataframe

```python
df.info()
```

Consultando los registros nulos, no se tiene ninguno

```python
nulls = df[df.isnull().any(axis=1)]
print(nulls)
```

Consultando los registros duplicados, no tiene ninguno

```python
dupli = df[df.duplicated()]
print(dupli)
```

Exportando el dataframe a un archivo `.csv`

```python
df.to_csv('../database/tetuan-pc.csv', index=False)
```
